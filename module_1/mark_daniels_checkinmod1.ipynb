### Linear regression 

1. What are the assumptions of linear regression? 

1. Variables have a linear relationship
2. Homoscedasticity
3. Normality of the model residuals (not the data necessarily)

2. Explain the reasoning for each assumption

1. If the relationship is not linear, we cannot accurately predict the relationship because there will be no way to make the prediction line match the curves of the 'real' data.
2. We need this to be true to avoid the circumstance where the variables are correlated in an uneven way so as to throw off our results/predictions. 
3. Normality of our residuals shows us that our model does not produce odd patterns in our errors (model residuals) which could indicate some sort of bias in our model (ie as one variable goes up the model becomes much less accurate/stable).

3. Explain the concept of correlation. What is the covariance? Why don't we use covariance? 

Correlation means that as the value of one variable goes up, it has a significant effect on the other variable.
Covariance is a measure of how the variables vary together.
We do not use covariance because it does not always scale well. Correlation normalizes and standardizes the measure to allow comparison between disparate variables.

4. Should we use the correlation to decide on variables to delete from our model?

We should not make that decision based on correlation alone. It can be a useful indicator if used properly.

5. What is R squared? How can it help us in comparing among models? 
What is the R squared of the baseline model?

R squared is a measure of how well our model predicts the variable(s) we are interested in predicting. It gives us another way of measure the error of our model. R squared of the baseline model will vary depending on how closely the data match the baseline. If the data are all equal to the mean, R squared will be 1.
