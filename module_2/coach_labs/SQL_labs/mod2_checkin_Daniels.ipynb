4. Use both the author table and books table to 
say: how many copies were sold by authors named David 


    a. how many books for each of the Davids
    
    
    b. how many books in total for all Davids

```python
c.execute('''SELECT * FROM author JOIN books ON author COUNT ([author])
            WHERE author LIKE 'David %' ;''').fetchall()
```

```python
c.execute('''SELECT author, SUM(num_copies) AS sold
            FROM author 
            LEFT JOIN books ON author.book = books.books
            GROUP BY author; ''')
```

```python
df = pd.DataFrame(c.fetchall())
df.columns = [i[0] for i in c.description]
df
```

```python
# group by authors named david
# sum num books by authors named david
# join on books

c.execute('''SELECT author AS authors_named_david, SUM(sold) AS sold_by_davids FROM
                    (SELECT author, SUM(num_copies) AS sold
                    FROM author 
                    LEFT JOIN books ON author.book = books.books
                    WHERE author LIKE 'David %'
                    GROUP BY author)
              WHERE author GROUP BY 'David'
              ; ''')
```

```python
df = pd.DataFrame(c.fetchall())
df.columns = [i[0] for i in c.description]
df
```
